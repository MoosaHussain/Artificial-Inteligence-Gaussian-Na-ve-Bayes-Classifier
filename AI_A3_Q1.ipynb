```python
#Author Moosa Hussain
#Q1 : Gaussian naive bayes Cllasifier 
#Load Data & Name the Columns 
import pandas as pd


#Step1 Preprocessing FeatureScaling of the dataframe 
from sklearn import preprocessing
def FeatureScaling(df):
    x = df.values #returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    df = pd.DataFrame(x_scaled)
    df.columns = ["F1","F2","F3","F4","F5","F6","F7","F8","F9","F10","F11","F12","F13","F14","F15","F16","F17","F18","F19","F20","F21","F22","Label"]
    return df


df = pd.read_excel (r'parktraining.xlsx')
df=FeatureScaling(df)

#Step2
#Prior Probability of Classes 
def Prior_Probability(e):
    zero= 0
    one = 0
    temp=[]
    for i in e :
        if i == 0.0 :
            zero=zero+1 
        if i==1.0:
            one=one+1
    p=len(e)        
    Prior_Zero = zero/p
    Prior_One  =one/p
    temp.append(Prior_Zero)
    temp.append(Prior_One)
    return(temp)
    
    
e=df['Label'] 
Prior=Prior_Probability(e)
#Step3

#Filter DataFrame LabelWise 
new= df['Label']  == 0.0
newo= df['Label'] == 1.0
Label_zero_df= df[new] 
Label_zero_Stats = Label_zero_df.describe()
Label_one_df= df[newo] 
Label_one_Stats = Label_one_df.describe()



#Step4 Calculation in Testing DataSet by using of Traing Dataset
import math

def Train_Zero_Label( feature , Train_Value):
    y= Train_Value - Label_zero_Stats[feature]['mean'] 
    py=-math.pow(y,2)
    std=Label_zero_Stats[feature]['std']
    pstd=2*pow(std,2)
    r=math.exp(py/pstd)
    t=math.sqrt(3.14*pstd)
    prob=r/t
    return prob

def Train_One_Label( feature , Train_Value):
    y= Train_Value - Label_zero_Stats[feature]['mean'] 
    py=-math.pow(y,2)
    std=Label_one_Stats[feature]['std']
    pstd=2*pow(std,2)
    r=math.exp(py/pstd)
    t=math.sqrt(3.14*pstd)
    prob=r/t
    return prob

#Load Testing DataSet
dataf = pd.read_excel (r'parktesting.xlsx')
tf=FeatureScaling(dataf)


def Calculate_Probability():
    name = ["F1","F2","F3","F4","F5","F6","F7","F8","F9","F10","F11","F12","F13","F14","F15","F16","F17","F18","F19","F20","F21","F22"]
    Zero_Prob = 1
    One_Prob = 1
    j=0
    Final_Zero=1
    Final_One=1
    
    value=[]
    Result_Array = []
    for k in range(len(tf)):
        for i in name :
        
            Zero_Prob = Zero_Prob * Train_Zero_Label(i,tf[i][k])
            One_Prob  = One_Prob * Train_One_Label(i,tf[i][k])
            j = j+1 
        
        
        Final_Zero = Zero_Prob * Prior[0]
        Final_One = One_Prob*Prior[1]
        Zero_Prob=1
        One_Prob=1
        value.append(Final_Zero)
        
        if(Final_Zero < Final_One):
            Result_Array.append(1)
        else:
            Result_Array.append(0)
    #print(value)
    return(Result_Array)

Result_A=Calculate_Probability()

tf['Label'] = Result_A
print("After Prediction Data Looklike ", tf.head())
def Accuracy():
    cf = pd.read_excel (r'parktesting.xlsx')
    Origional_Data= cf[1.000000]
    Predicted_Data=tf['Label']
    
    check = 0
    o=0
    for i in range(len(Origional_Data)):
        
        if Origional_Data[check] == Predicted_Data[ check ]:
            o=o+1
            
        check= check+1

    u=len(Origional_Data)
    acc =(o/u) *100
    return(acc)

print("Accuracy = " , Accuracy())

```
